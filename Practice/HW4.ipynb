#### c. (1 point) Based on your answer from part b), revise your import process from part a) so that you can use the dictionary to create a more appropriately formatted DataFrame. Demonstrate your improved process by displaying (printing) your modified dictionary format and previewing the first 10 rows of your resultant DataFrame.

```python
# revise your import process from part a
inc100_c={}
## open csv and build dictionary
with open('inc5000_2017.csv','r')as csv_file:
    ## read csv as dictionary
    data=csv.DictReader(csv_file, delimiter=",")
    ## 5 empty dinctionaries to store values corresponding to 5 keys
    dict_name = {}
    dict_found = {}
    dict_industry ={}
    dict_metro={}
    dict_workers={}
    ## loop to extract values
    for row in data:
        ## only extract top 100 values
        if int(row['rank']) <101:
            ## matching keys and values
            dict_inc_name={int(row['rank']): row['company']}
            dict_inc_found={int(row['rank']): row['founded']}
            dict_inc_industry={int(row['rank']): row['industry']}
            dict_inc_metro={int(row['rank']): row['metro']}
            dict_inc_workers={int(row['rank']): row['workers']}
            dict_name.update(dict_inc_name)
            dict_found.update(dict_inc_found)
            dict_industry.update(dict_inc_industry)
            dict_metro.update(dict_inc_metro)
            dict_workers.update(dict_inc_workers)
        inc100_c['name']= dict_name
        inc100_c['founded']= dict_found
        inc100_c['industry']= dict_industry
        inc100_c['metro']= dict_metro
        inc100_c['workers']= dict_workers
inc100_c
```

#### h. (2 points) Explore the types of industries in some of the top 20 cities. Specifically, write a function sum_industry that takes in a DataFrame (df) and a city (default to New York City), and returns the frequency counts of the industry column. Use the function to summarize the industry concentration for New York City, Atlanta, Chicago, and Washington (DC). What observations can you make about the industry concentration in each of these markets?

```python
df_h = df_d_adj.pivot_table(index='city',columns='industry', values='company', aggfunc='count',fill_value=0)
```

```python
def sum_industry(dataframe=df_h, city='New York City'):
    print('industry concentration for', city, ':')
    print(dataframe.loc[city].sort_values(ascending=False))
```

Most of companies in New York City concentrate on Advertising and Marketing industry. Also, Business Products & Services and IT Services are other two main industries that companies in New York stayed in. 

```python
# summarize the industry concentration for New York City
sum_industry(df_h)
```

Atlanta shows its competitive advantage on IT Services industries. Numbers of companies in Advertising, and Business Products and Services are also quite big.

```python
# summarize the industry concentration for Atlanta
sum_industry(df_h,'Atlanta')
```

Chicago has almost same numbers of companies in Business Products & Service, IT Service, Real Estate and Software industries, and relatively more companies in Advertising & Marketing industry.

```python
# summarize the industry concentration for Chicago
sum_industry(df_h,'Chicago')
```

In Washington, number of companies in Government Services are significantly higher than that in other industries.

```python
# summarize the industry concentration for Washington (DC)
sum_industry(df_h,'Washington')
```
