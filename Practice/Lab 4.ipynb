```python
import pandas as pd
```

```python
solar_system = pd.read_html('https://en.wikipedia.org/wiki/List_of_natural_satellites', match='Image')[0]
solar_system
```

* Drop Image and Ref(s)

```python
solar_system.drop(['Image','Ref(s)'], axis=1, inplace=True)
solar_system
```

* Reorder the columns

```python
solar_system = solar_system[['Planet','Discovery year','Name','Numeral','Discovered by','Mean radius (km)','Semi-major axis (km)','Sidereal period (d) (r = retrograde)','Notes']]
solar_system
```

* Replace all "—" to NaN

```python
import numpy as np
```

```python
solar_system.replace({'—': np.nan}, inplace=True)
solar_system
```

* Delete numbers after '±' ,and '±' as well as '~' for 'Mean radius (km)' and 'Semi-major axis (km)' then turn it into 'int' type
* Delete '–' and the number after '–'

```python
solar_system['Mean radius (km)'] = solar_system['Mean radius (km)'].str.split(expand=True)
solar_system['Semi-major axis (km)'] = solar_system['Semi-major axis (km)'].str.split(expand=True)
solar_system['Mean radius (km)'] = solar_system['Mean radius (km)'].str.split("–", expand=True)
solar_system['Mean radius (km)'] = solar_system['Mean radius (km)'].str.replace('±','')
solar_system['Mean radius (km)'] = solar_system['Mean radius (km)'].str.replace('~','')
solar_system['Semi-major axis (km)'] = solar_system['Semi-major axis (km)'].str.replace(',','')
solar_system['Mean radius (km)'] = solar_system['Mean radius (km)'].astype('float', copy=False)
solar_system['Semi-major axis (km)'] = solar_system['Semi-major axis (km)'].astype('float', copy=False)
```

* Hierarchical indexing (Planet, Discovery year)

```python
solar_system = solar_system.set_index(keys=['Planet','Discovery year']).sort_index(level=(0,1))
```
